**<a id="#">Task 4.1.1.</a>** Load the image `data/astronaut.png` and show it:

**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
v_mask = numpy.array([[ -1,  -1,  -1],
                       [ 0,  0,  0],
                       [1, 1, 1]])
h_mask = numpy.array([[-1, 0, 1],
                       [-1, 0, 1],
                       [-1, 0, 1]])
def prewitt_h(img_in):
    result = numpy.zeros(img_in.shape)
    n = (3 - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        i, j = p
        if i in range(n, img_in.shape[0] - n):
            if j in range(n, img_in.shape[1] - n):
                neighborhood_sum = 0
                for q in numpy.ndindex(3, 3):
                    neighborhood_sum = neighborhood_sum + img_in[i-n+q[0], j-n+q[1]]*h_mask[q[0],q[1]]
                result[i,j] = neighborhood_sum*(1/6)
    return result
    
def prewitt_v(img_in):
    result = numpy.zeros(img_in.shape)
    n = (3 - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        i, j = p
        if i in range(n, img_in.shape[0] - n):
            if j in range(n, img_in.shape[1] - n):
                neighborhood_sum = 0
                for q in numpy.ndindex(3, 3):
                    neighborhood_sum = neighborhood_sum + img_in[i-n+q[0], j-n+q[1]]*v_mask[q[0],q[1]]
                result[i,j] = neighborhood_sum*(1/6)
    return result
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
gx=prewitt_h(img)
plt.imshow(gx, 'gray')
plt.colorbar()
```

```python
gy=prewitt_v(img)
plt.imshow(gy, 'gray')
plt.colorbar()
```

## Sobel filters (bonus)

**<a id="#">Task 4.3.1.</a>** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical 3&times;3 Sobel filtering, respectively:

```python
v_s_mask = numpy.array([[ -1,  -2,  -1],
                       [ 0,  0,  0],
                       [1, 2, 1]])
h_s_mask = numpy.array([[-1, 0, 1],
                       [-2, 0, 2],
                       [-1, 0, 1]])
def sobel_h(img_in):
    result = numpy.zeros(img_in.shape)
    n = (3 - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        i, j = p
        if i in range(n, img_in.shape[0] - n):
            if j in range(n, img_in.shape[1] - n):
                neighborhood_sum = 0
                for q in numpy.ndindex(3, 3):
                    neighborhood_sum = neighborhood_sum + img_in[i-n+q[0], j-n+q[1]]*h_s_mask[q[0],q[1]]
                result[i,j] = neighborhood_sum*(1/8)
    return result
    
def sobel_v(img_in):
    result = numpy.zeros(img_in.shape)
    n = (3 - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        i, j = p
        if i in range(n, img_in.shape[0] - n):
            if j in range(n, img_in.shape[1] - n):
                neighborhood_sum = 0
                for q in numpy.ndindex(3, 3):
                    neighborhood_sum = neighborhood_sum + img_in[i-n+q[0], j-n+q[1]]*v_s_mask[q[0],q[1]]
                result[i,j] = neighborhood_sum*(1/8)
    return result
```
